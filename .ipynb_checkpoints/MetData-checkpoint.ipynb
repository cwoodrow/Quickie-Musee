```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('https://media.githubusercontent.com/media/metmuseum/openaccess/refs/heads/master/MetObjects.csv',low_memory=False)
gender_df = pd.read_csv('name_gender_dataset.csv',low_memory=False)

print(df.head(10))
print(df.columns)
```

```python
df_filtered = df.drop(['Tags AAT URL','Tags Wikidata URL', 'Link Resource', 
               'Is Highlight', 'Is Timeline Work', 'Is Public Domain', 'Culture', 'Period', 'Dynasty', 
               'Reign', 'Portfolio', 'Repository', 'Subregion', 'Locale', 'Locus', 'Excavation',
               'River', 'Classification', 'Rights and Reproduction', 'Object Wikidata URL',
               'Metadata Date', 'Object ID', 'Object Number', 'Constituent ID', 'Artist Role',
               'Artist Prefix', 'Artist Suffix', 'Artist End Date', 'Artist ULAN URL', 'Artist Wikidata URL'], axis=1)

print(df_filtered.head(10))
df_filtered.shape
```

```python
df_filtered['Department'].value_counts()
```

```python
df_filtered = df_filtered[df_filtered['Department'] == 'Modern and Contemporary Art'] 
```

```python
df_filtered[df_filtered['Tags'].str.contains("Female Nudes")==True].value_counts()

df_filtered[df_filtered['Tags'].str.contains("Male Nudes")==True]
```

```python
female_nudes = len(df_filtered[df_filtered['Tags'].str.contains("Female Nudes")==True])
male_nudes = len(df_filtered[df_filtered['Tags'].str.contains("Male Nudes")==True])
total_nudes = female_nudes + male_nudes

print("female_nudes : " + "%.2f" % (female_nudes/total_nudes * 100) + "%"
      + ", male_nudes : " + "%.2f" % (male_nudes/total_nudes * 100) + "%")


print("female_nudes : " + "%.2f" % (female_nudes)
      + ", male_nudes : " + "%.2f" % (male_nudes))
```

```python
df_filtered['Tags'].value_counts()
```

```python
df_filtered['Artist Display Name'].value_counts()

#Artist First Name 
df_filtered['Artist Name Split']=(df_filtered['Artist Display Name'].str.split(' '))
df_filtered['Artist First Name']=np.where(df_filtered['Artist Name Split'].isnull() | len(df_filtered['Artist Name Split']) == 0, '', df_filtered['Artist Name Split'].str[0])
df_filtered = df_filtered.drop(['Artist Name Split'], axis=1)
df_filtered
```

```python
gender_df = gender_df.drop_duplicates(subset=['Name'])
gender_df
```

```python
df_gendered = df_filtered.merge(gender_df, left_on='Artist First Name',right_on='Name', how='left')
df_gendered
```

```python
df_gendered = df_gendered.drop(['Artist First Name', 'Name', 'Count', 'Probability'], axis=1)
df_gendered
```

```python
df_gendered['Gender'] = df_gendered['Gender'].fillna('Unknown')
df_gendered['Gender'].value_counts(dropna=False)
```

```python
df_exposed = df_gendered
df_exposed['Gender'].value_counts(dropna=False)
```

```python
df_exposed['AccessionYear'].astype('str').min()

df_exposed['BeforeOrAfter1985'] = pd.cut(df_exposed['AccessionYear'].astype('str'), ['1900', '1985', '2025'], 3, labels=["Before 1985", "After 1985"])

df_exposed.groupby('BeforeOrAfter1985',observed=True)['Gender'].value_counts()
```

```python
df_gendered['Exposed'] = df_gendered['Gallery Number'].notnull()
df_exposed.groupby(['Gender', 'BeforeOrAfter1985'],observed=True)['Exposed'].value_counts().reset_index().to_csv('met-artwork-gendered.csv')
```

```python
df_gendered.groupby(['AccessionYear','Gender'])['Exposed'].value_counts().reset_index().to_csv('met-artworks-per-accession-year.csv')
```
